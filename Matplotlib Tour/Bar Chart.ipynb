```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
from matplotlib import style
```

```python
classes=["Python","AI","ML","R","DS"]
class_1=[30,10,20,25,10]
class_2=[40,5,20,20,10]
class_3=[35,5,30,15,15]
```

# 1.bar(): it is used for plotting the bar graph.

```python
plt.bar(classes,class_1)
```

# 2.barh() : used for plotting the bar graph horizontally.

```python
plt.barh(classes,class_1)
```

# 3.width: use to assign width of bar chart.

```python
plt.bar(classes,class_1,width=0.4)
```

```python
plt.bar(classes,class_1,width=0.4,color="red")
```

# 4.edge : use to align the bar chart on center or edge.

```python
plt.bar(classes,class_1,width=0.4,color="red",align='edge')
```

# 5.egdecolor : use to color the edge of bar.

```python
plt.bar(classes,class_1,width=0.4,color="red",align='edge',edgecolor='black')
```

# 6.linewidth: use to assign width of edge of bar chart.

```python
plt.bar(classes,class_1,width=0.4,color="red",align='edge',edgecolor='black',linewidth=5)
```

# 7.alpha: gives opacity to the edges of bar chart.

```python
plt.bar(classes,class_1,width=0.4,color="red",align='edge',edgecolor='black',linewidth=5,alpha=0.6)
```

# 8.linestyle: style for edge.

```python
plt.bar(classes,class_1,width=0.4,color="red",align='edge',edgecolor='black',linewidth=5,alpha=0.6,linestyle='--')
```

# 9.visible: visibility of the bar chart.

```python
plt.bar(classes,class_1,width=0.4,color="red",align='edge',edgecolor='black',linewidth=5,alpha=0.6,linestyle='--',visible=False)
```

# 10.figure size.

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class_1,width=0.4,color="red",align='edge',edgecolor='black',linewidth=5,alpha=0.6,linestyle='--')
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class_1,width=0.4,color="red",align='edge',edgecolor='black',linewidth=5,alpha=0.6,linestyle='--')
plt.title("bar chart",fontsize=30)
plt.xlabel("Subjects",fontsize=28)
plt.ylabel("No. of students",fontsize=28)
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class_1,width,color="red",align='edge')
plt.bar(classes_index + width,class_2,width,color="black",align='edge')
plt.title("bar chart",fontsize=30)
plt.xlabel("Subjects",fontsize=28)
plt.ylabel("No. of students",fontsize=28)
plt.show()
```

# for getting labels as x co-ordinates on numbers use xticks.It replaces the number with labels that we want to give.

```python
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class_1,width,color="red",align='edge')
plt.bar(classes_index + width,class_2,width,color="black",align='edge')
plt.xticks(classes_index+width,classes)
plt.title("bar chart",fontsize=30)
plt.xlabel("Subjects",fontsize=28)
plt.ylabel("No. of students",fontsize=28)
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class_1,width,color="red",align='edge')
plt.bar(classes_index + width,class_2,width,color="black",align='edge')
plt.xticks(classes_index+width,classes,rotation=20)
plt.title("bar chart",fontsize=30)
plt.xlabel("Subjects",fontsize=28)
plt.ylabel("No. of students",fontsize=28)
plt.show()
```
