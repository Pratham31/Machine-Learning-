# It uses image library from matplotlib for reading other image formats.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# How to show image using matplotlib imshow function?

```python
img=mpimg.imread("pie_char.png")
```

```python
img
```

# - It converts the image into an array of three dimensions.

```python
type(img)
```

```python
img.shape
```

```python
img.ndim
```

# 1.imshow - use to show image.

```python
plt.imshow(img)
```

# 2.show() - for removing extra info. other than image.

```python
plt.imshow(img)
plt.show()
```

# 3.axis("off") - use to remove axis from image. 

```python
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img)
plt.show()
```

# 4.colorbar() - Use to colorbar.

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img,cmap="Accent")
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img,cmap="hot")
plt.colorbar()
plt.show()
```

```python
sinle_channel=img[:,:,0]
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(sinle_channel,cmap="hot")
plt.colorbar()
plt.show()
```
