# 3.For scatter plot you must use x and y co-ordinate from dataset inorder to    plot. 

```python
x=df_playstore['Rating']
y=df_playstore['Reviews']
```

# 4.Actual Plotting of Scatter plot.

```python
plt.scatter(x,y)
plt.title("Google Play Store Scatter plot")
plt.xlabel('Ratings')
plt.ylabel('Reviews')
```

# 5.color of points : color or c.

```python
plt.scatter(x,y,c='black')
plt.title("Google Play Store Scatter plot")
plt.xlabel('Ratings')
plt.ylabel('Reviews')
```

# 6.increase size of plotting - figure(figsize).

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='black')
plt.title("Google Play Store Scatter plot")
plt.xlabel('Ratings')
plt.ylabel('Reviews')
```

# 7.marker - use to style the plotting co-ordinates.

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='black',marker='*')
plt.title("Google Play Store Scatter plot")
plt.xlabel('Ratings')
plt.ylabel('Reviews')
```

# 8.size of marker - s.

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='black',marker='*',s=100)
plt.title("Google Play Store Scatter plot")
plt.xlabel('Ratings')
plt.ylabel('Reviews')
```

# 9.alpha - for transparency of markers.

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='black',marker='*',s=100,alpha=0.5)
plt.title("Google Play Store Scatter plot")
plt.xlabel('Ratings')
plt.ylabel('Reviews')
```

# 10.linewidth - edge width of marker.

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='black',marker='*',s=100,alpha=0.5,linewidth=10)
plt.title("Google Play Store Scatter plot")
plt.xlabel('Ratings')
plt.ylabel('Reviews')
```

# 11.edgecolor - color of edge of markers.

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='black',marker='*',s=100,alpha=0.5,linewidth=10,edgecolor='red')
plt.title("Google Play Store Scatter plot")
plt.xlabel('Ratings')
plt.ylabel('Reviews')
```

# 12.verts - verts is same as marker.In order to use verts remove marker. 

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='black',s=100,alpha=0.5,linewidth=10,edgecolor='red',verts='_')
plt.title("Google Play Store Scatter plot")
plt.xlabel('Ratings')
plt.ylabel('Reviews')
```

# 13.Merging of two Scatter plots.

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="black",marker="+",s=100,alpha=0.5,linewidths=10,edgecolors="g")
plt.scatter(x,df_playstore["Installs"],c="red",marker="o",s=100,alpha=0.5,linewidths=10,edgecolors="yellow")
plt.title("Google Play Store Scatter plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews and Installs")
```
