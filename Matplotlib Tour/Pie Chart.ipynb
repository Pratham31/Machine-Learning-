```python
import matplotlib.pyplot as plt
```

```python
classes=['Python','R','Machine Learning','Artificial Intelligence','Data Science']
class_stud=[49,15,35,25,30]
```

# 1.Plotting Simple pie-chart.

```python
plt.pie([1])
plt.show()
```

# 2.Plotting pie-chart from given list of class_stud.

```python
plt.pie(class_stud)
plt.show()
```

# 3.Giving Labels to pie-chart.

```python
plt.pie(class_stud,labels=classes)
plt.show()
```

# 4.explode - use to slice one portion of pie-chart.

```python
explode=[.03,0,.1,0,0]
plt.pie(class_stud,labels=classes,explode=explode)
plt.show()
```

# 5.Assigning colors to the pir-chart.

```python
colors=['c','b','r','y','g']
explode=[.03,0,.1,0,0]
plt.pie(class_stud,labels=classes,explode=explode,colors=colors)
plt.show()
```

# 6.autopct - Shows percentage of different sections.

```python
colors=['c','b','r','y','g']
explode=[.03,0,.1,0,0]
plt.pie(class_stud,labels=classes,explode=explode,colors=colors,autopct="%.2f%%")
plt.show()
```

# 7.shadow - gives shadow to pie-chart if True.

```python
colors=['c','b','r','y','g']
explode=[.03,0,.1,0,0]
plt.pie(class_stud,labels=classes,explode=explode,colors=colors,autopct="%.2f%%",shadow=True)
plt.show()
```

# 8.radius - it can be used to increase the radius of pie-chart.

```python
colors=['c','b','r','y','g']
explode=[.03,0,.1,0,0]
plt.pie(class_stud,labels=classes,explode=explode,colors=colors,autopct="%.2f%%",shadow=True,radius=1.8)
plt.show()
```

# 9.startangle - use to start our pie-chart from specified angle in anti-clockwise.

```python
colors=['c','b','r','y','g']
explode=[.03,0,.1,0,0]
plt.pie(class_stud,labels=classes,explode=explode,colors=colors,autopct="%.2f%%",shadow=True,radius=1.8,startangle=90)
plt.show()
```

# 10.textprops - Using textprops we can give desired font size.

```python
colors=['c','b','r','y','g']
explode=[.03,0,.1,0,0]
plt.pie(class_stud,labels=classes,explode=explode,colors=colors,autopct="%.2f%%",shadow=True,radius=1.8,
        startangle=90,textprops={"fontsize":12})
plt.show()
```

# OR 

```python
colors=['c','b','r','y','g']
explode=[.03,0,.1,0,0]
textprops={"fontsize":12}
plt.pie(class_stud,labels=classes,explode=explode,colors=colors,autopct="%.2f%%",shadow=True,radius=1.8,
        startangle=90,textprops=textprops)
plt.show()
```

# 11.Exploring all paramters of Pie-chart.

```python
wedgeprops={"linewidth":4,"width":1,"edgecolor":"black"}
plt.pie(
    class_stud,
    explode=[.07,0,0,0,0],
    labels=classes,
    colors=['c','b','r','y','g'],
    autopct="%.2f%%",
    pctdistance=0.8,
    shadow=True,
    labeldistance=1.1,
    startangle=90,
    radius=1.5,
    counterclock=True,
    wedgeprops=wedgeprops,
    textprops={"fontsize":24},
    center=(0,0),
    frame=True,
    rotatelabels=True)
plt.show()
```

# 12.legend() - showing labels corresponds to color.

```python
colors=['c','b','r','y','g']
explode=[.03,0,.1,0,0]
textprops={"fontsize":12}
plt.pie(class_stud,labels=classes,explode=explode,colors=colors,autopct="%.2f%%",shadow=True,radius=1.8,
        startangle=90,textprops=textprops)
plt.legend()
plt.show()
```
