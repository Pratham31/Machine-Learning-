```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_stu_age=np.random.randint(18,45,(100))
py_stu_age=np.random.randint(15,40,(100))
```

```python
print(ml_stu_age)
print(py_stu_age)
```

# 1.Histogram - hist(): 

```python
plt.hist(ml_stu_age)

plt.title('ML Student Age Histogram')
plt.xlabel("student age")
plt.ylabel("no. of students")
plt.show()
```

# 2.bins: is use for x co-ordinate

```python
plt.hist(ml_stu_age,bins=[15,20,25,30,35,40,45])

plt.title('ML Student Age Histogram')
plt.xlabel("student age")
plt.ylabel("no. of students")
plt.show()
```

# 3.rwidth: For separating the histograms based on percentage.

```python
plt.hist(ml_stu_age,bins=[15,20,25,30,35,40,45],rwidth=.8)

plt.title('ML Student Age Histogram')
plt.xlabel("student age")
plt.ylabel("no. of students")
plt.show()
```

# 4.histtype : Takes type of histogram.

```python
#by default histtype is bar
plt.hist(ml_stu_age,bins=[15,20,25,30,35,40,45],rwidth=.8,histtype='step')

plt.title('ML Student Age Histogram')
plt.xlabel("student age")
plt.ylabel("no. of students")
plt.show()
```

# 5.Orientation : shows histogram horizontal or vertical.

```python
plt.hist(ml_stu_age,bins=[15,20,25,30,35,40,45],rwidth=.8,orientation='horizontal')

plt.title('ML Student Age Histogram')
plt.xlabel("student age")
plt.ylabel("no. of students")
plt.show()
```

# 6.color: For changing the color of histogram.

```python
plt.hist(ml_stu_age,bins=[15,20,25,30,35,40,45],rwidth=.8,color='red')

plt.title('ML Student Age Histogram')
plt.xlabel("student age")
plt.ylabel("no. of students")
plt.show()
```

```python
plt.hist(ml_stu_age,bins=[15,20,25,30,35,40,45],rwidth=.8,color='red',label='ml student')

plt.title('ML Student Age Histogram')
plt.xlabel("student age")
plt.ylabel("no. of students")
plt.legend()
plt.show()
```

```python
plt.hist(py_stu_age,bins=[15,20,25,30,35,40,45],rwidth=.8,color='black',label='py student')

plt.title('Python Student Age Histogram')
plt.xlabel("student age")
plt.ylabel("no. of students")
plt.legend()
plt.show()
```

```python
plt.hist(ml_stu_age,bins=[15,20,25,30,35,40,45],rwidth=.8,color='red',label='ml student')
plt.hist(py_stu_age,bins=[15,20,25,30,35,40,45],rwidth=.8,color='black',label='py student',orientation='horizontal')
plt.title('ML Student Age Histogram')
plt.xlabel("student age")
plt.ylabel("no. of students")
plt.legend()
plt.show()
```

# 7.Merging of two graphs : Just write points,color,labels in list.

```python
plt.hist([ml_stu_age,py_stu_age],bins=[15,20,25,30,35,40,45],rwidth=.8,color=['red','black'],label=['ml student','Py student'])
#plt.hist(py_stu_age,bins=[15,20,25,30,35,40,45],rwidth=.8,color='black',label='py student')
plt.title('ML Student Age Histogram')
plt.xlabel("student age")
plt.ylabel("no. of students")
plt.legend()
plt.show()
```

# 8.figsize : Takes width and height as input & shows graph accord.

```python
plt.figure(figsize=(16,9))

plt.hist([ml_stu_age,py_stu_age],bins=[15,20,25,30,35,40,45],rwidth=.8,color=['red','black'],orientation='vertical',label=['ml student','Py student'])
#plt.hist(py_stu_age,bins=[15,20,25,30,35,40,45],rwidth=.8,color='black',label='py student')

plt.title('ML Student Age Histogram')
plt.xlabel("student age")
plt.ylabel("no. of students")
plt.legend()
plt.show()
```
