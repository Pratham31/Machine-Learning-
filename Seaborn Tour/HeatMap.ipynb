```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr2D=np.linspace(1,5,12).reshape(4,3)
arr2D
```

```python
glob_df=pd.read_csv("D:\\Machine Learning\\Who_is_responsible_for_global_warming.csv")
glob_df
```

## Here We have to take only numerical Value and on X-axis we have Years and On Y-axis we have Country Name .

```python
glob_df=glob_df.drop(columns=['Country Code','Indicator Name','Indicator Code'],axis=1).set_index('Country Name')
glob_df.head(6)
```

## Increase size by figsize function.

```python
plt.figure(figsize=(16,9))
sns.heatmap(glob_df)
```

## 1.vmax - Use to set y-max value in color bar of heatmap.
## 2.vmin - Use to set y-min value in color bar of heatmap.

```python
plt.figure(figsize=(16,9))
sns.heatmap(glob_df,vmin=0,vmax=21)
```

## 3.Use to colormap for better understanding.

```python
plt.figure(figsize=(16,9))
sns.heatmap(glob_df,vmin=0,vmax=21,cmap='coolwarm')
```

## 4.annot - (Annotation) Takes boolean Values,if true it shows the grid value.

```python
plt.figure(figsize=(16,9))
sns.heatmap(glob_df,vmin=0,vmax=21,annot=True,cmap='coolwarm')
```

## 5.annot_kws - We can use annotation Keyword to set different parameters.
### i) fontsize - Assigns Font size.
### ii) fontstyle - Use to assign style for font.
### iii) color - Color of the font.
### iv) alpha - Transparency of font.
### v) rotation - Rotation of font (Horizontal or Vertical).
### vi) verticalalignment - Alignment of Fonts .
### vii) backgroundcolor - Color for Background .


```python
plt.figure(figsize=(16,9))
annot_kws={'fontsize':10,
          'fontstyle':'italic',
          'color':'k',
          'alpha':.9,
          'rotation':'vertical',
          'verticalalignment':'center',
          'backgroundcolor':'w',}
sns.heatmap(glob_df,vmin=0,vmax=21,annot=True,cmap='coolwarm',annot_kws=annot_kws)
```

## 6.linewidth - Use to give linewidth to matrix grid.

```python
plt.figure(figsize=(16,9))
sns.heatmap(glob_df,vmin=0,vmax=21,annot=True,cmap='coolwarm',linewidth=4)
```

## 7.linecolor - Use to color the linewidth.

```python
plt.figure(figsize=(16,9))
sns.heatmap(glob_df,vmin=0,vmax=21,annot=True,cmap='coolwarm',linewidth=4,linecolor='black')
```

## 8.cbar(color bar) - Again It takes Boolean values if True,it shows color bar and if False,Hides Color bar.

```python
plt.figure(figsize=(16,9))
sns.heatmap(glob_df,vmin=0,vmax=21,annot=True,cmap='coolwarm',linewidth=4,cbar=False)
```

## 9.xticks and yticks - Takes boolean value.Use to show or hide X and Y axis Labels.

```python
plt.figure(figsize=(16,9))
sns.heatmap(glob_df,vmin=0,vmax=21,annot=True,cmap='coolwarm',linewidth=4,xticklabels=False,yticklabels=False)
```

## 10.cbar_kws - color bar Keyword.
### i) orientation - can be Horizontal or Vertical.
### ii) shrink - Use to assign size for color bar.
### iii) extend - takes min or max i.e makes tip(head) on min or max value(depends on Input).
### iv) extendfrac - Use to extend the tip(head) of color bar. More value more tip(head) will extended.
### v) ticks - Values on Color bar axis.
### vi) drawedges - Draws edges on colorbar .

```python
plt.figure(figsize=(16,9))
cbar_kws={'orientation':'horizontal',
         'shrink':1,
         'extend':'min',
         'extendfrac':0.1,
         'ticks':np.arange(0,20),
         'drawedges':True}
sns.heatmap(glob_df,cbar_kws=cbar_kws)
```

## 11.set function - Use to set title,X-axis Label and Y-axis Label.

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(glob_df)
ax.set(title="Heatmap of who is repsonsible for Global Warming",
      xlabel="CO2 emissions Per Year",
      ylabel="Country Name")
```

## 12.font Scaling - Use to scale all the labels on Heatmap.

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(glob_df)
ax.set(title="Heatmap of who is repsonsible for Global Warming",
      xlabel="CO2 emissions Per Year",
      ylabel="Country Name")
sns.set(font_scale=1)
```

## Correlation Heatmap - It is statistical measure of the relationship between two variables (X,Y).

```python
glob_df.corr()
```

## Plotting of correlation

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(glob_df.corr(),annot=True,linewidth=3)
ax.tick_params(size=10,color='w',labelsize=10,labelcolor='black')
plt.title("Heatmap of who is repsonsible for Global Warming",fontsize=25)
plt.show()
```

## Importing dataset from sklearn.

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset=load_breast_cancer()
cancer_dataset
```

## Conversion of dataset into dataframe

```python
cancer_df=pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
                      columns=np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
plt.figure(figsize=(25,25))
ax=sns.heatmap(cancer_df.corr(),annot=True,linewidth=3)
ax.tick_params(size=10,color='w',labelsize=10,labelcolor='black')
plt.title("Heatmap of Breast Cancer",fontsize=25)
plt.show()
```
