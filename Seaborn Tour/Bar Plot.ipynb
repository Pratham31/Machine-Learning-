```python
import seaborn as sns
import numpy as np
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

## Bar Plot -

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill)
```

## hue - use to categorize the bar plot.

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill,hue=tips_df.sex)
```

## bar plot shorthand Property - Use data="dataframe_name". By doing so, there is no need to write Dataframe_Name.column ex. "tips_df.day" in function.

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df)
```

```python
order = ['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order)
```

```python
hue_order=['Female','Male']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,hue_order=hue_order)
```

## estimator - Use to set max value on Y-axis w.r.t. Y label

```python
hue_order=['Female','Male']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,hue_order=hue_order,estimator=np.max)
```

## ci(Confidence Interval) - use to specify length of error bar.

```python
hue_order=['Female','Male']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,ci=20)
```

```python
hue_order=['Female','Male']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,hue_order=hue_order,estimator=np.max,n_boot=2)
```

## For changing the orientation interchange the x and y co-ordinates means put categorical value on Y-axis.

```python
hue_order=['Female','Male']
sns.barplot(x='total_bill',y='day',hue='sex',data=tips_df,hue_order=hue_order,estimator=np.max)
```

```python
hue_order=['Female','Male']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,hue_order=hue_order,estimator=np.max,palette="hot")
```

## errwidth - Use to specify Error Bar Width.

```python
hue_order=['Female','Male']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,hue_order=hue_order,estimator=np.max,palette="hot",errwidth=12)
```

## errcolor - Use to specify color of Error Bar.

```python
hue_order=['Female','Male']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,hue_order=hue_order,estimator=np.max,
            palette="hot",errwidth=12,errcolor="green")
```

## dodge - Use to show two bars into one.

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,dodge=False)
```

## kwargs(Keyword Arguments) - Passing as a dictionary.

```python
kwargs={'alpha':0.9,'linestyle':':','linewidth':5,'edgecolor':'black'}
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,**kwargs
           )
```

## kwargs(Keyword Arguments) - Passing as a Parameters.

```python

sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,alpha=0.9,linestyle=':',linewidth=5,edgecolor='black')
```

## Axes - Use to assign title, X and Y axis Label.

```python

ax=sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,alpha=0.9,linestyle=':',linewidth=5,edgecolor='black')
ax.set(title="Bar Plot ",
      xlabel="Days",
      ylabel="Total Bill")
```
