```python
import seaborn as sns
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

# Histogram Plotting - use distplot() as method.

## - It also shows line plot known as Kernal Density Estimiter(KDE).

```python
sns.distplot(tips_df["size"])
```

```python
sns.distplot(tips_df["tip"])
```

```python
sns.distplot(tips_df["total_bill"])
```

 ## 1.bins is used to divide the data into the parts.Following Exapmle will divide data into 100 bars.

```python
sns.distplot(tips_df["total_bill"],bins=100)
```

## 2.(hist=False) - Use to remove Histogram only shows Line Plot.

```python
sns.distplot(tips_df["total_bill"],hist=False)
```

## 3.(kde=False) - Use to remove Line plot only shows Histogram.

```python
sns.distplot(tips_df["total_bill"],kde=False)
```

## 4.(rug=True) - Use to show Points on X-Axis.

```python
sns.distplot(tips_df["total_bill"],rug=True)
```

## 5.Changing the orienation of Plotting.

```python
sns.distplot(tips_df["total_bill"],rug=True,vertical=True)
```

## 6.(fit=norm) - Use to show normalized Line Plot.

```python
from scipy.stats import norm
sns.distplot(tips_df["total_bill"],fit=norm)
```

## 7.(color="red") - Changing the color of hist plot

```python
sns.distplot(tips_df["total_bill"],color="red")
```

##  8.axlabel="axis-name" - Use to give name to the X-axis.

```python
sns.distplot(tips_df["total_bill"],axlabel="Total Bill")
```

## 9.Use label to describe information regarding Plot and title using matplotlib.

```python
import matplotlib.pyplot as plt
sns.distplot(tips_df["total_bill"],color="red",label="Total Bill")
plt.legend()
```

```python
sns.distplot(tips_df["total_bill"],color="red",label="Total Bill")
plt.title("Histogram Plotting")
plt.legend()
plt.show()
```

## set() - Use to set background Grid.

```python
sns.set()
sns.distplot(tips_df["total_bill"],label="Total Bill")
plt.title("Histogram Plotting")
plt.legend()
plt.show()
```

# bins Exploration

## -- bins - Here we can define range of co-ordinates that we want to distribute on X-axis and plot histogram Accordingly.

```python
plt.figure(figsize=(16,9))

bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df["total_bill"],label="Total Bill",bins=bins)
plt.xticks(bins)
plt.show()
```

## hist_kws(keyword Argument) - Can be used to give no. of parameters at a time.

```python
sns.distplot(tips_df["total_bill"],label="Total Bill",
             hist_kws={'color':'red'})
```

```python
sns.distplot(tips_df["total_bill"],label="Total Bill",
             hist_kws={'color':'red','edgecolor':'yellow','linewidth':3,'linestyle':'--'})
```

## kde_kws - Can be used to no.of parameters at a time.

```python
plt.figure(figsize=(16,8))
sns.distplot(tips_df["total_bill"],label="Total Bill",
             hist_kws={'color':'red','edgecolor':'yellow','linewidth':3,'linestyle':'--'},
            kde_kws={'color':'black','linewidth':4,'linestyle':'--'})
```
