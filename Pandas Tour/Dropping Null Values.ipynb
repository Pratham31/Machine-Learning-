```python
import pandas as pd
```

```python
df=pd.read_csv('D:\\Machine Learning\\newdata.csv')
df
```

# dropna() : simply removes NaN values

```python
df.dropna()
```

# removing NaN values from row (axis=0)

```python
df.dropna(axis=0)
```

# removing NaN values from column (axis=1)

```python
df.dropna(axis=1)
```

# Demonstration of how

```python
df.dropna(how='any') #by default removes null values from row 
```

```python
df.dropna(axis=1,how='any')
```

# how='all' - drops row or column only if contains all NaN values in row or col. 

# thresh=

** thresh gives output based on your input argument in thresh function as follows:
 if you give 2 as parameter then it simply gives output containing atleast two not null value in row

```python
df.dropna(thresh=3)
```

```python
df.dropna(thresh=2)
```

# subset

```python
df.dropna(subset=['name']) #drops entire row containing NaN values corresponding to column name
```
