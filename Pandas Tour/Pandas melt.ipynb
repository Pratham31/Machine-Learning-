# melt - is used to transform or reshape data

```python
import pandas as pd
```

```python
df1=pd.read_csv('D:\\Machine Learning\\movies.csv')
df1
```

# After using melt function it assigns variables to column name and values to values in respective column

```python
pd.melt(df1)
```

```python
pd.melt(df1,id_vars=['Genre'])
```

# value_vars - generally used to unpivot table(converts the format(many columns) into long format(few columns but many rows) )

```python
pd.melt(df1,id_vars=['year'],value_vars=['Genre'])
```

#  ^ in above output we will get year wise data or values of Genre

# *We can change value name and variable name with help of  value_name,var_name

```python
pd.melt(df1,id_vars=['year'],value_vars=['Genre'],var_name='Var',value_name='Valu')
```
