```python
import pandas as pd
```

```python
df=pd.read_csv('D:\\Machine Learning\\newdata.csv')
df
```

# fillna()

```python
#filling na values with default
df.fillna(value='default')
```

```python
#filling na values with 0
df.fillna(value=0)
```

```python
#filling na values in names column with 0
df.fillna({'name':0})
```

```python
#filling na values in branch column with none
df.fillna({'branch':'none'})
```

```python
#method is way to fill na values
#ffill(forward fill) = means it fills nan values with respect to its previous records
df.fillna(method='ffill')
```

```python
#bfill(backward fill) = means it fills nan values with respect to its backward records
df.fillna(method='bfill')
```

```python
df.fillna(method='pad')#same as ffill
```

```python
df.fillna(method='bfill',axis=0)
```

```python
df.fillna(value=0,limit=1)
```

```python
df.fillna(0,limit=1)
```

```python
df.fillna(5,inplace=True)#it only modifies the dataframes
df
```
