```python
import pandas as pd
```

```python
df=pd.read_csv('D:\\\Machine Learning\\newdata.csv')
df
```

```python
#na_values takes parameters that will be converted into NaN
df=pd.read_csv('D:\\\Machine Learning\\newdata.csv',na_values=['no','no value','not appl'])
df
```

```python
#for applying NaN value to the specific column only
df=pd.read_csv('D:\\\Machine Learning\\newdata.csv',na_values={'name':'no value'})
df
```

```python
#na_values takes parameters dictionary
df=pd.read_csv('D:\\\Machine Learning\\newdata.csv',na_values={'name':['no value','no','not appl']})
df
```

# keep_default_na

```python
#if it is false then keeps the records as it is without assigning NaN
#by default True
df=pd.read_csv('D:\\\Machine Learning\\newdata.csv',keep_default_na=False)
df
```

# na_filter

```python
#It is used to increase performance
#use it only when csv doesnt contain any NaN value
df=pd.read_csv('D:\\\Machine Learning\\newdata.csv',na_filter=False)
df
```

# isnull()

```python
#isnull simply prints true is there is no record and false if contains record
df=pd.read_csv('D:\\\Machine Learning\\newdata.csv')
df.isnull()
```

# sum of null values

```python
df.isnull().sum()
```

# Total Null values

```python
df.isnull().sum().sum()
```
