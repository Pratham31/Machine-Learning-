```python
import pandas as pd
```

```python
df1=pd.read_csv('D:\\Machine Learning\\movies.csv')
df1
```

# Pivot_table() - It generally gives Aggregate output depends on Argument.

# index - if you provide only index then it will give you aggregate mean of values

```python
df1.pivot_table(index='Genre') #resultant will be aggregate result of each same cells in Genre
```

# Aggregate Function - aggfunc='sum' - Returns Aggregate sum

```python
df1.pivot_table(index='Genre',aggfunc='sum')
```

# aggfunc='max' | aggfunc='min' - returns max,min values resp.

```python
df1.pivot_table(index='Genre',aggfunc='max')
```

# columns - shows cells according to given column values 

```python
df1.pivot_table(index='Genre',columns='year')
```

# aggfunc='count' - gives aggregate count of index argument

```python
df1.pivot_table(index='Genre',aggfunc='count')
```

# fill_value : It generally fills NaN value 

```python
df1.pivot_table(index='Genre',columns='year',fill_value=0)
```

# margins : it adds given rows w.r.t column by default it gives mean value where there is no use of aggfunc

```python
df1.pivot_table(index='Genre',columns='year',margins=True)
```
