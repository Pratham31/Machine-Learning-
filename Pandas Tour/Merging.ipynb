```python
import pandas as pd
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4,5],
                 'class':[10,10,11,12,12]})
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4,5],
                 'name':['A','B','C','D','E']})
df2
```

# - merge function()

```python
df3=df1.merge(df2)
df3
```

# - df1 on Left Side and df2 on Right side

```python
#or
pd.merge(df1,df2)
```

```python
#Merging on ID column
pd.merge(df1,df2,on='ID')
```

# - df2 on Left Side and df1 on Right side

```python
pd.merge(df2,df1)
```

# if ID is not same in both dataframe then it will show only same records

```python
df3=pd.DataFrame({'ID':[1,2,3,4],
                 'class':['A','B','C','D']})
df3
```

```python
pd.merge(df1,df3)
```

# how argument--default inner(intersection between two dataframes)

```python
pd.merge(df1,df3,how='left')
```

```python
pd.merge(df1,df3,how='right')
```

```python
pd.merge(df1,df3,how='outer')
```

```python
pd.merge(df1,df3,how='outer',indicator=True)
```

# ^ indicator use for which dataframe contains value in case of join

```python
pd.merge(df1,df3,left_index=True,right_index=True)
```

# - suffixes in DataFrame 

# * Note - For suffix column name of two df's should be same

```python
pd.merge(df1,df3,on='ID',suffixes=('_left','_right')) 
```
